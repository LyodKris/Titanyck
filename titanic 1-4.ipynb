```python
import re
import numpy as np
import pandas as pd

def toFixed(numObj, digits=0):
    return f"{numObj:.{digits}f}"

data = pd.read_csv('titanic.csv', delimiter=',')
mail_cnt = 0
femail_cnt = 0
alive = 0
frst_class_pas = 0
middl_age = 0
pas_amount = len(data)
dict_femail_name = {}

#Количество мущжчин и женщин        
mail_cnt = data['Sex'] == "male"
mail_cnt = mail_cnt.sum()
femail_cnt = data['Sex'] == "female"
femail_cnt = femail_cnt.sum()

#Процент пассажиров первого класса
frst_class_pas = data["Pclass"] == 1
frst_class_pas = frst_class_pas.sum()
frst_class_prsnt = 100 * frst_class_pas / pas_amount

#Процент пассажиров выжило
alive = data["Survived"] == 1
alive = alive.sum()
alive_prsnt = 100 * alive / pas_amount

#Средний возраст пассажиров и медиана возрастов
medium_age = data["Age"].mean()
median_age = data.median()

#Корреляция Пирсона между признаками SibSp и Parch
cor_pirs = data.SibSp.corr(data.Parch)

#Самое частое жен. имя на борту
for i in range(pas_amount):
    name = ""
    if data.Sex[i] == 'female':
        if len(data.Name[i].split('Ms. ')) == 2:
            name = data.Name[i].split('Ms. ')
        if len(data.Name[i].split('Mrs. ')) == 2:
            name = data.Name[i].split('Mrs. ')
        if len(data.Name[i].split('Miss. ')) == 2:
            name = data.Name[i].split('Miss. ')
    if name == "":
        continue
    name = name[1].split(' ')
    if name[0][0] == "(":
        name = name[0][1:]
    else:
        name = name[0]
    if name[-1] == ")":
        name = name[:len(name) - 1]
    if name in dict_femail_name:
        dict_femail_name[name] += 1
    else:
        dict_femail_name[name] = 1
max = ""
for i in dict_femail_name.keys():
    if max == "":
        max = i
    if dict_femail_name[i] > dict_femail_name[max]:
        max = i

print(f"Пассажиров \n  мужского пола: \t\t\t{mail_cnt}\n  женского пола: \t\t\t{femail_cnt}\n  пол не указан:\t\t\t{len(data) - mail_cnt - femail_cnt}")
print(f"Пассажиров 1го класса:\t\t\t{toFixed(frst_class_prsnt, 2)}%")
print(f"Пассажиров выжило:\t\t\t{toFixed(alive_prsnt, 2)}%")
print(f"Средний возраст пассажиров:\t\t{toFixed(medium_age, 0)}")
print(f"Медиана возрастов пассажиров:\t\t{toFixed(medium_age, 0)}")
print(f"Корреляция Пирсона между SibSp и Parch:\t{toFixed(cor_pirs, 2)}")
print(f'Самое частое жен. имя на борту {max}, оно встречается {dict_femail_name[max]} раз')

```
